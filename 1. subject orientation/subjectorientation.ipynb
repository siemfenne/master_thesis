```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('HadCRUT5_summary_series.csv')
df.head()
```

```python
df['Time'] = pd.to_datetime(df['Time'])
df
```

```python
plt.plot(df['Time'], df['Anomaly (deg C)'], linewidth=0.5)
plt.xlabel('Time (years)')
plt.ylabel('Anomaly (deg C) relative to 1961-1990')
plt.savefig('tempseries.png', bbox_inches='tight', dpi=300)
plt.show()
```
